```python
import pandas as pd
import folium
import json

# Carga de datos
saps = pd.read_csv('Odata.xlsm - SAPS.csv')
barrios = pd.read_csv('Odata.xlsm - Barrios.csv')
barrios_populares = pd.read_csv('bdgis_RENABAP - bdgis_RENABAP.csv')

# Crear mapa centrado en Corrientes
m = folium.Map(location=[-27.469213, -58.830635], zoom_start=12, tiles=None)

# Añadir polígonos para delimitar barrios no populares en azul
for _, row in barrios.iterrows():
    geo_json = json.loads(row['st_asgeojson'])
    folium.GeoJson(
        geo_json,
        name=row['nombre_barrio'],
        tooltip=row['nombre_barrio'],
        style_function=lambda x: {
            'color': 'blue',
            'weight': 2,
            'fillOpacity': 0.3
        }
    ).add_to(m)

# Añadir polígonos para delimitar barrios populares en rojo (encima de los azules)
for _, row in barrios_populares.iterrows():
    geo_json = json.loads(row['path'])
    folium.GeoJson(
        geo_json,
        name=row['nombre_barrio'],
        tooltip=row['nombre_barrio'],
        style_function=lambda x: {
            'color': 'red',
            'weight': 2,
            'fillOpacity': 0.3
        }
    ).add_to(m)

# Añadir marcadores al mapa para los SAPs
for _, row in saps.iterrows():
    lat = row['lat']
    lng = row['lng']
    name = row['nombre']
    folium.Marker([lat, lng], tooltip=name).add_to(m)

# Guardar el mapa en un archivo HTML
m.save('mapa.html')

m
```

```python
densidad = gpd.read_file('vw_densidad_de_poblacion.csv')

# Convierte la columna 'zona' de formato GeoJSON a tipo geométrico
densidad['geometry'] = densidad['zona'].apply(lambda x: shape(json.loads(x)))

# Convertir el DataFrame en un GeoDataFrame
gdf = gpd.GeoDataFrame(densidad, geometry='geometry')

# Dibuja el mapa de calor basado en la columna densidad_poblacion
fig, ax = plt.subplots(figsize=(15, 15)) # Aumenta el tamaño del mapa

# Dibujar el mapa de calor
gdf.plot(column='densidad_poblacion', 
         cmap='OrRd', 
         legend=True, 
         ax=ax,
         legend_kwds={'orientation': "horizontal"}) # Hacer la barra de colores horizontal

# Título y límites
ax.set_title('Densidad Poblacional de Corrientes', fontsize=16)
ax.set_xlim(gdf.geometry.bounds.minx.min(), gdf.geometry.bounds.maxx.max()) # Ajustar los límites en X
ax.set_ylim(gdf.geometry.bounds.miny.min(), gdf.geometry.bounds.maxy.max()) # Ajustar los límites en Y

plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Agrupamos por fecha y SAPS, y luego sumamos la cantidad de consultas
agrupado_por_fecha_saps = consultas.groupby(['fecha', 'saps']).consulta_cantidad.sum().reset_index()

# Ahora graficamos
fig = px.line(agrupado_por_fecha_saps, x='fecha', y='consulta_cantidad', color='saps', title='Cantidad de pacientes por día separado por SAPS')
fig.show()
```

```python

# Agrupamos por fecha y SAPS, y luego sumamos la cantidad de consultas
agrupado_por_fecha_saps = consultas.groupby(['fecha', 'saps']).consulta_cantidad.sum().reset_index()

# Filtramos para que la fecha comience en 2020
agrupado_por_fecha_saps = agrupado_por_fecha_saps[agrupado_por_fecha_saps['fecha'] >= '2020-01-01']

# Creamos un gráfico de barras para cada SAPS
saps_unicos = agrupado_por_fecha_saps['saps'].unique()
for saps in saps_unicos:
    data_saps = agrupado_por_fecha_saps[agrupado_por_fecha_saps['saps'] == saps]
    fig = px.bar(data_saps, x='fecha', y='consulta_cantidad', title=f'Cantidad de pacientes por día en {saps}')
    fig.show()
```
